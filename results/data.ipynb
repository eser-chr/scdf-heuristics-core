```python
import struct
import numpy as np

with open("tuning_ga_objectives_over_time.bin", "rb") as f:
    rows = struct.unpack('Q', f.read(8))[0]
    data = []
    for _ in range(rows):
        cols = struct.unpack('Q', f.read(8))[0]
        row = np.frombuffer(f.read(cols * 8), dtype=np.float64)
        data.append(row)
    data = np.array(data)
```
